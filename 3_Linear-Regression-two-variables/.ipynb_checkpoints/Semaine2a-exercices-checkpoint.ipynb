## Calcul d'une première prédiction

### Initialisez theta en un vecteur de deux valeurs à zéro

```python
theta = np.array([0,0])
print(theta)
print(type(theta))
```

### Écrivez une fonction _predict_ qui prend en argument une population (x) ainsi que les parametres theta et prédit le profit (y) associé

```python
def predict(X, theta):
    I = np.ones((len(X)))
    M = np.array([I,X])
    H = theta.dot(M)
    return(H)
#     return theta[0] + np.dot(theta[1],X)
```

```python
theta = np.array([-3.6303,1.1664])
#theta = np.array([0.,0.])
H = predict(X, theta)
# print(H)
# print(y)
# print(H - y)

import matplotlib.pyplot as plt
print(plt.scatter(X,H))
print(plt.scatter(X,y))
```

### Écrivez une fonction _fit_ qui prend en arguments le vecteur X et le vecteur y des données d'entraînement et renvoie le vecteur de paramètres _theta_ qui a été appris

```python
def fit(X, y, theta, alpha, num_iters):
    # Initialiser certaines variable utiles
    m = len(X)
#     I_bis = np.ones(len(X))
    
    # Boucler sur le nombre d'itérations
    
    #for i in range(1, n):
        # Effectuer une itération de descente du gradient (i.e. on update theta une fois)
    for i in range(1, num_iters):
        H = predict(X,theta)
#         theta[0] = theta[0] - 1 / m * alpha * (H-y).dot(I_bis.T)
#         theta[1] = theta[1] - 1 / m * alpha * (H-y).dot(X.T)
        theta[0] = theta[0] - 1 / m * alpha * np.sum((H-y))
        theta[1] = theta[1] - 1 / m * alpha * np.sum((H-y) * X)
    return theta
```

### Lancez l'apprentissage en appelant la fonction _fit_ et en prenant bien soin de récupérer le résultat de *theta* à la fin!! Vous devriez obtenir des valeurs autour de [-3.6303, 1.1664]

Pour commencer, on fixera alpha à 0.01 et num_iters à 1500

```python
alpha = 0.01
num_iters = 1500
theta = np.array([0.,0.])

theta = fit(X, y, theta, alpha, num_iters)
print(theta)
```

## Fonction de coût

Maintenant voyons comment resserrer notre analyse de l'algorithme et calculons le coût (ou la perte, ou l'erreur) à chaque itération.

### Définissez la fonction de coût de votre modèle

```python
def cost(X, y, theta):
    m = len(X)
    H = predict(X,theta)
    J = 1 / m / 2 * np.sum((H-y) * (H-y))
    return(J)
```

### Testez-la avec theta = [0,0]   Vous devriez obtenir environ 32.07

```python
theta = np.array([0.,0.])
print(cost(X, y, theta))
```

### Maintenant avec theta = [-1,2]. Vous devriez obtenir environ 54.24

```python
theta = np.array([-1.,2.])
print(cost(X, y, theta))
```

### Copiez le code de votre fonction _fit_ et ajoutez-y un appel à la fonction _cost_, à chaque itération. Vous stockerez vos résultats dans une liste nommée J_history, que vous retournerez avec *theta* à la fin de la fonction

Et oui, en Python, une fonction peut retourner plus qu'une variable!!

```python
def fit_with_cost(X, y, theta, alpha, num_iters):
   # À compléter
    J_history = 0
    
    num_iters_list = 0
    theta_0_list = theta[0]
    theta_1_list = theta[1]
    
    m = len(X)
    I_bis = np.ones(len(X))
    
    # Boucler sur le nombre d'itérations
    for i in range(0, num_iters):
        num_iters_list = np.append(num_iters_list, i)
        J_history = np.append(J_history, cost(X, y, theta))
        H = predict(X,theta)
        theta[0] = theta[0] - alpha * 1 / m * 2 * (H-y).dot(I_bis.T)
        theta[1] = theta[1] - alpha * 1 / m * 2 * (H-y).dot(X.T)
        theta_0_list = np.append(theta_0_list, theta[0])
        theta_1_list = np.append(theta_1_list, theta[1])
        
#     return theta, J_history
    return theta, J_history, theta_0_list, theta_1_list, num_iters_list
```

### Appelez la fonction en récupérant les valeurs de theta et J_history

```python
# D'abord on réinitialise theta à zéro
theta = np.zeros(2)
```

```python
# Lancez l'entraînement avec votre nouvelle fonction
theta = np.zeros(2)
num_iters = 1500

J_history = fit_with_cost(X, y, theta, alpha, num_iters)[1]
theta_0_list = fit_with_cost(X, y, theta, alpha, num_iters)[2]
theta_1_list = fit_with_cost(X, y, theta, alpha, num_iters)[3]
num_iters_list = fit_with_cost(X, y, theta, alpha, num_iters)[4]

print(J_history)
print(theta_0_list)
print(theta_1_list)
```

```python
theta = np.zeros(2)
num_iters = 1500
alpha = 0.01

J_history = fit_with_cost(X, y, theta, alpha, num_iters)[1]
theta_0_list = fit_with_cost(X, y, theta, alpha, num_iters)[2]
theta_1_list = fit_with_cost(X, y, theta, alpha, num_iters)[3]
num_iters_list = fit_with_cost(X, y, theta, alpha, num_iters)[4]

print(J_history)
print(theta_0_list)
print(theta_1_list)

fit = plt.figure()
ax_0 = plt.axes()
ax_0.set(title='J en fct de theta 0', xlabel ='theta 0',
           ylabel='J')
ax_0.plot(theta_0_list, J_history, color = 'purple')

print(ax_0)

fit = plt.figure()
ax_1 = plt.axes()
ax_1.set(title='J en fct de theta 1', xlabel ='theta 1',
           ylabel='J')
ax_1.plot(theta_1_list, J_history, color = 'lightblue')
print(ax_1)

fit = plt.figure()
ax_3 = plt.axes()
ax_3.set(title='J en fct de num_iters', xlabel ='theta 1',
           ylabel='J')
ax_3.plot(num_iters_list, J_history, color = 'orange')
print(ax_1)
```
